```python
def getNum(*, chars: str) -> int:
    freqs = [0] * 9
    nums = int(''.join([x for x in chars if x.isdigit()]))
    while nums:
        num = nums % 10
        freqs[num - 1] += 1
        nums //= 10
    return freqs.index(max(freqs)) + 1
```

```python
getNum(chars = text)
```
